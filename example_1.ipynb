```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
```

```python
oecd_bli=pd.read_csv("oecd_bli_2015.csv",thousands=',')
gdp_per_capita=pd.read_csv("gdp_per_capita.csv",thousands=',',delimiter='\t',encoding='latin1',na_values="n/a")
```

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd = oecd_bli[oecd_bli["INEQUALITY"] == "TOT"].copy()
    oecd = oecd.pivot(index="Country", columns="Indicator", values="Value")

    gdp = gdp_per_capita.rename(columns={"2015": "GDP per capita"})
    gdp = gdp[["Country", "GDP per capita"]].set_index("Country")

    full_country_stats = pd.merge(
        left=oecd, right=gdp, left_index=True, right_index=True
    )
    full_country_stats.sort_values(by="GDP per capita", inplace=True)

    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(
        set(range(len(full_country_stats))) - set(remove_indices)
    )
    return full_country_stats[["GDP per capita", "Life satisfaction"]].iloc[keep_indices]
```

```python
country_stats=prepare_country_stats(oecd_bli,gdp_per_capita)
x=np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"].to_numpy()]
```

```python
country_stats.plot(kind='scatter',x="GDP per capita",y="Life satisfaction")
plt.show()
```

```python
model=sklearn.linear_model.LinearRegression()
```
